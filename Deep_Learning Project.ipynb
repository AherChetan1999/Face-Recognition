# Face Recognition

```python
import face_recognition
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy
```

```python
fig = plt.figure(figsize=(20,4))
img = cv2.imread("dd.jpg")
image = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
face_loc = face_recognition.face_locations(image)
num_face = len(face_loc)
```

```python
plt.imshow(image)
ax = plt.gca()
for face in face_loc :
    top , right , bottom , left = face
    x,y,w,h = left , top, right , bottom
    print("top :{} , bottom:{}, left:{} , right :{}" . format(y,h,x,w))
    rect = Rectangle((x,y) , w-x , h-y , fill= False , color='red')
    ax.add_patch(rect)
plt.show()
```

```python
img1 = cv2.imread("dd.jpg")
image1 = cv2.cvtColor(img1 , cv2.COLOR_BGR2RGB)
img2= cv2.imread("dd2.jpg")
image2 = cv2.cvtColor(img2 , cv2.COLOR_BGR2RGB)
img3 = cv2.imread("amitabh.jpg")
image3 = cv2.cvtColor(img3 , cv2.COLOR_BGR2RGB)
```

```python
face_demo1 = face_recognition.face_encodings(image1)[0]
face_demo2 = face_recognition.face_encodings(image2)[0]
face_demo3 = face_recognition.face_encodings(image3)[0]
```

### Create Database

```python
known_face_encodings =[face_demo1 , face_demo2 ,face_demo3]
```

```python
img = cv2.imread("amitabh.jpg")
un_image = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(un_image)
unknown_face_encodings = face_recognition.face_encodings(un_image)
```

```python
from scipy.spatial import distance
import numpy as np
plt.imshow(un_image)

for unknown_face_encoding in unknown_face_encodings:

    results=[]
    for known_face_encoding in known_face_encodings:
        d = distance.euclidean(known_face_encoding , unknown_face_encoding)
        results.append(d)
    threshold=0.6
    results = np.array(results) <= threshold
    
name="unknown"
    
if results[0]:
        name = "face Donald Trump"
elif results[1]:
        name = "face Pm Modi"
elif results[2]:
        name = "face Amitabh Bacchan"

print(f"Found {name} in the photo")

    
    
```
